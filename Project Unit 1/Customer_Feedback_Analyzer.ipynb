##Pre-trained transformer-based sentiment analysis model provided by Hugging Face

```python
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

##Testing with one sentence

```python
result = classifier("The product quality is very good and I love it")
print(result)
```

##Testing with multiple reviews

```python
reviews = [
  "This phone is amazing",
  "Battery life is terrible",
  "I love this product",
  "Worst service ever",
  "Good value for money"
]

for r in reviews:
    print(r, "->", classifier(r))
```
