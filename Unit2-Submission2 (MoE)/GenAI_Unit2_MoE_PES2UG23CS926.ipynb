```python
import os
from groq import Groq

# Set your Groq API key
import getpass
os.environ["GROQ_API_KEY"] = getpass.getpass("Enter your Groq API key: ")

client = Groq(api_key=os.environ["GROQ_API_KEY"])

BASE_MODEL = "llama-3.1-8b-instant"
```

```python
MODEL_CONFIG = {
    "technical": {
        "system_prompt": (
            "You are a Technical Support Expert. "
            "Be precise, analytical, and code-focused. "
            "Provide step-by-step debugging help. "
            "If relevant, include corrected code snippets."
        ),
        "temperature": 0.7,
    },
    "billing": {
        "system_prompt": (
            "You are a Billing Support Specialist. "
            "Be empathetic and polite. "
            "Explain refund policies clearly. "
            "Guide users through next steps calmly."
        ),
        "temperature": 0.7,
    },
    "general": {
        "system_prompt": (
            "You are a friendly customer support assistant. "
            "Answer casually and helpfully."
        ),
        "temperature": 0.7,
    },
    "crypto": {
        "system_prompt": (
            "You are a cryptocurrency information assistant."
        ),
        "temperature": 0.3,
    }
}
```

```python
def route_prompt(user_input: str):
    """
    Classifies user input into:
    technical, billing, general, or crypto.

    Returns:
        category (str)
        raw_response (str)
    """

    routing_prompt = (
        "You are an intent classifier.\n"
        "Classify the text into exactly ONE of these categories:\n"
        "technical\n"
        "billing\n"
        "general\n"
        "crypto\n\n"
        "Return ONLY the category name. No explanation.\n\n"
        f"Text: {user_input}"
    )

    response = client.chat.completions.create(
        model=BASE_MODEL,
        messages=[
            {"role": "system", "content": "You are a strict classifier."},
            {"role": "user", "content": routing_prompt}
        ],
        temperature=0
    )

    raw_output = response.choices[0].message.content.strip().lower()

    # Enforce valid categories
    if raw_output not in MODEL_CONFIG:
        print("[Router Warning] Invalid category returned. Defaulting to general.")
        return "general", raw_output

    return raw_output, raw_output
```

```python
def fetch_bitcoin_price():
    """
    Mock tool function simulating real-time API call.
    """
    return "The current price of Bitcoin is $51,240 (mock data)."
```

```python
def process_request(user_input: str) -> str:
    """
    Production-style MoE Orchestrator
    """

    print("\n--- ROUTING PHASE ---")
    category, raw = route_prompt(user_input)

    print(f"[Router Output] → {raw}")
    print(f"[Final Category] → {category}")

    print("\n--- EXPERT ACTIVATION PHASE ---")

    # Tool-based expert
    if category == "crypto":
        print("[Expert Type] → Tool-Based Expert")
        return fetch_bitcoin_price()

    # LLM-based expert
    expert_config = MODEL_CONFIG[category]

    print(f"[Expert Type] → LLM Expert ({category.upper()})")

    response = client.chat.completions.create(
        model=BASE_MODEL,
        messages=[
            {"role": "system", "content": expert_config["system_prompt"]},
            {"role": "user", "content": user_input}
        ],
        temperature=expert_config["temperature"]
    )

    return response.choices[0].message.content
```

```python
user_input = input("Enter your query: ")
result = process_request(user_input)

print("\n[Final Expert Response]")
print(result)
```
